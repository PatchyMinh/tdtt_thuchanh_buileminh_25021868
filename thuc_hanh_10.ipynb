```python
# W9A1

def merge(lst1, lst2):
    i = 0 # lst1
    j = 0 # lst2
    lstOut = []
    while i < len(lst1) and j < len(lst2):
        if lst1[i] > lst2[j]:
            lstOut.append(lst2[j])
            j += 1
        else:
            lstOut.append(lst1[i])
            i += 1
    if i < len(lst1):
        lstOut += lst1[i:]
    else:
        lstOut += lst2[j:]
    return lstOut
lst1 = list(input().split())
lst2 = list(input().split())
print(" ".join(merge(lst1, lst2)))
```

```python
# W9A2
m = int(input()) # cột
n = int(input()) # hàng
matrix = []
for _ in range(n):
    lstTemp = list(map(int, input().split()))
    matrix.append(lstTemp)
print(matrix)
k = int(input())
left = min(matrix[0])
right = max(matrix[-1])
countTemp = 0
while left <= right:
    guessTemp = (left + right) // 2
    for i in range(n):
        for j in range(m):
            if matrix[i][j] <= guessTemp:
                countTemp += 1
    if countTemp < k:
        left = guessTemp + 1
    else:
        right = guessTemp - 1
    countTemp = 0
print(left) 
```

```python
# W9A4
def Modifiable(strIn, strCheck):
    i = len(strIn)
    j = len(strCheck)
    if i == j == 0:
        return True
    if i == 0 and j > 0:
        return False
    if i != 0 and j == 0:
        for item in strIn:
            if item.isupper():
                return False
        return True
    if strIn[i-1].isupper():
        if strIn[i-1] == strCheck[j-1]:
            return Modifiable(strIn[:i-1], strCheck[:j-1])
        else:
            return False
    else:
        option_delete = Modifiable(strIn[:i-1], strCheck)
        option_transform = False
        if strIn[i-1].upper() == strCheck[j-1]: # Nhớ dùng j-1
            option_transform = Modifiable(strIn[:i-1], strCheck[:j-1])
        return option_delete or option_transform
q = int(input())
for _ in range(q):
    strIn = input()
    strCheck = input()
    if Modifiable(strIn, strCheck):
        print("YES")
    else:
        print("NO")
```

```python
# W9A5

lst1 = list(map(int, input().split()))
lst2 = list(map(int, input().split()))
 
i = len(lst1)
j = len(lst2)
def check_exact(i, j):
    if i == -1: 
        return True       
    if j == -1: 
        return False      
    if lst1[i] != lst2[j]: 
        return False 
    return check_exact(i - 1, j - 1)
def is_subset(j):
    if j < len(lst1) - 1:
        return False
    if lst1[-1] == lst2[j]:
        if check_exact(len(lst1) - 1, j):
            return True
    return is_subset(j - 1)

if is_subset(len(lst2) - 1):
    print("YES")
else:
    print("NO")
```

```python
# W9A6

n = int(input())
strIn = input()
lstIn = list(map(int, strIn.split()))
def fac(n):
    product = 1
    if n == 0 or n == 1:
        return product
    else:
        for i in range(2, n + 1):
            product *= i
    return product
def permutation(n):
    result = []
    def backtrack(path):
        if len(path) == n:
            result.append(" ".join(str(item) for item in path))
        for item in range(1, n + 1):
            if item in path:
                continue
            path.append(item)
            backtrack(path)
            path.pop()
    backtrack([])
    return result
lstOut = permutation(n)
indexOut = lstOut.index(strIn)
print(lstOut[(indexOut + 1) % fac(n)])

# Thuật trên chạy trâu không tốt.
# Dự kiến làm lại:
# B1: tìm điểm sao cho A[i] < A[i + 1]
# B2: đổi A[i] với inf của dãy bên phải
# B3: rồi sắp xếp lại dãy trước A[i]
# B4: (nếu hoán vị là hoán vị cuối thì đảo ngược xâu)
```

```python
# W9A7

def canPlaceFlowers(flowerbed, k):
    length = len(flowerbed)
    def value(i, k): # index
        if k <= 0:
            return True
        if i >= length:
            return False
        if flowerbed[i] == "1":
            return value(i + 2, k)
        if flowerbed[i] == "0":
            left_empty = (i == 0) or (flowerbed[i-1] == "0")
            right_empty = (i == length - 1) or (flowerbed[i+1] == "0")
            if left_empty and right_empty:
                return value(i + 2, k - 1)
            else:
                return value(i + 1, k)
    return value(0, k)

flowerbed = list(input().split())
k = int(input())
print(canPlaceFlowers(flowerbed, k))
    
```

```python
# W9A8

def is_path_crossing(moves):
    x_rect = 0
    y_rect = 0
    move_offset = {"R": (1, 0), "L": (-1, 0), "U": (0, 1), "D": (0, -1)}
    list_coordinate = {(0, 0)}
    for item in moves:
        dx, dy = move_offset[item]
        x_rect += dx
        y_rect += dy
        if (x_rect, y_rect) not in list_coordinate:
            list_coordinate.add((x_rect, y_rect))
        else:
            return True
    return False
moves = input()
print(is_path_crossing(moves))


```

```python
# W9A9
def Median(nums):
    a, b, c, d, e = nums[0], nums[1], nums[2], nums[3], nums[4]
    
    if a > b:
        a, b = b, a
    
    if c > d:
        c, d = d, c
    
    if a > c:
        a, c = c, a
        b, d = d, b 
    
    if b > e:
        b, e = e, b
    
    if b < c:
        if c < e:
            return c
        else:
            return e
    else:
        if b < d:
            return b
        else:
            return d
    
```

```python
# W9A10
def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()
    j = 0
    res = 0
    # i - houses, j - heaters
    for house in houses:
        while j < len(heaters) - 1 and abs(heaters[j+1] - house) <= abs(heaters[j] - house):
            j += 1
        res = max(res, abs(heaters[j] - house))
    return res

houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))
findRadius(houses, heaters)
```
