```python
# W10A3

try:
    n = int(input())
    sumTemp = 0
    setTemp = set()
    while n != 1:
        while n > 0:
            sumTemp += (n % 10)**2
            n = n // 10
        n = sumTemp
        sumTemp = 0
        if n not in setTemp:
            setTemp.add(n)
        else:
            break
    if 1 in setTemp:
        print("YES")
    else:
        print("NO")
except ValueError:
    print("Không nhập được chữ cái")

        
```

```python
# W10A10
try:
    lstIn = list(map(int, input().split()))
    lstTemp = lstIn[1:]
    setTemp = set()
    for item in lstTemp:
        if item not in setTemp:
            setTemp.add(item)
        else:
            raise ValueError
    print(sum(lstTemp))
except ValueError:
    print("Mang khong hop le")
```

```python
# W10A11
import re

def is_valid_regex(pattern):
    try:
        re.compile(pattern)
        return True
    except re.error:
        return False

try:
    t_str = input() 
    if t_str:
        t = int(t_str)
        for _ in range(t):
            s = input()
            if is_valid_regex(s):
                print("True")  # Sample output ghi True/False
            else:
                print("False")
except (EOFError, ValueError):
    pass
```
