```python
# W11A1

def reverse (path: str) -> int:
    with open(path, "r") as file:
        content = file.read()
    number_str = content.strip()

    return int(number_str[::-1])

with open("number.txt", "w") as f:
    f.write(input())

result = reverse("number.txt")
print(result)
```

```python
# W11A2

def canWinNim (path: str) -> bool:
    with open(path, "r") as f:
        num = int(f.read())
    if num % 4 == 0:
        return False
    else:
        return True

with open("nim_num.txt", "w") as f:
    f.write(input())

result = canWinNim("nim_num.txt")
print(result)
```

```python
# W11A4

def areAnagrams(path1: str, path2: str) -> bool:
    with open(path1, "r") as f1, open(path2, "r") as f2:
        name1 = f1.read().strip().lower()
        name2 = f2.read().strip().lower()
    nameOut1 = ""
    nameOut2 = ""
    for item in name1:
        if item != " ":
            nameOut1 += item
    for item in name2:
        if item != " ":
            nameOut2 += item
    if sorted(nameOut1) == sorted(nameOut2):
        return True
    else:
        return False

path1 = input()
path2 = input()

print(areAnagrams(path1, path2))
```

```python
def count(path: str, character: str) -> int:
    with open(path, "r") as f:
        content = f.read()
    count = 0
    for item in content:
        if item == character:
            count += 1
    return count

fileDir = input()
charCheck = input()
count(fileDir, charCheck)
```

```python
def countCharacter(path: str, letter: str) -> int:
    with open(path, "r") as f:
        content = f.read().lower()
    count = 0
    for item in content:
        if item == letter:
            count += 1
    return count

dirIn = input()
letter = input()
countCharacter(dirIn, letter)
```

```python
# W11A8
f = open("fibonacci.txt", "w")
def Fibonacci(n, lstTemp = []):
    for i in range(0, n):
        if i == 0:
            lstTemp.append("a")
        elif i == 1:
            lstTemp.append("b")
        else:
            lstTemp.append(lstTemp[i - 1] + lstTemp[i - 2])
    return lstTemp
n = int(input())
for item in Fibonacci(n):
    f.write(item + '\n')
f.close()
```

```python
# W11A9

def grade10(path: str):
    lstTemp = []
    with open(path, "r") as f:
        content = f.readlines()
        for item in content:
            temp = item.rstrip().split(",")
            lstTemp.append(temp)
    for item in lstTemp[1:]:
        if int(float(item[5])) == 10:
            print(f"{item[0]}, {item[1]} {item[2]}")
    return
grade10("tabulardata.txt")
```
