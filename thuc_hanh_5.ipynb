```python
# W5A1

def Max_of_Two(a, b):
    if a > b:
        return a
    else:
        return b
x, y = map(int, input("Nhập hai số nguyên để trả ra số nguyên lớn hơn:").split())
Max_of_Two(x, y)

```

```python
# W5A2

def swapInt(a, b):
    return (b, a)
x, y = map(int, input("Nhập hai số nguyên để đổi chỗ cho chúng, cách bằng một dấu cách: ").split())
swapInt(x, y)
```

```python
# W5A3

def isPrime(intCheck):
    if intCheck == 0 or intCheck == 1:
        return False
    else:
        i = 2
        while i**2 <= intCheck:
            if intCheck % i == 0:
                return False
            else:
                i += 1
        return True

print(isPrime(int(input("Nhập số nguyên n, trả về True nếu là số nguyên tố, False nếu không phải:"))))
```

```python
# W5A4

def isPerfect(intCheck):
    sumDivisors = 0
    for i in range(1, intCheck):
        if intCheck % i == 0:
            sumDivisors += i
    if sumDivisors == intCheck:
        return True
    else:
        return False

isPerfect(int(input("Nhập số nguyên n, trả về True nếu là số hoàn hảo, False nếu không phải:")))
```

```python
# W5A5

def findNumber(lst, numFind):
    index = 0
    for num in lst:
        index += 1
        if num == numFind:
            return index
    return -1

lstNum = list(map(int, input("Nhập một dãy số cách nhau bằng một dấu cách: ").split()))
numFind = int(input("Nhập số nguyên bạn cần tìm: "))
findNumber(lstNum, numFind)

        
```

```python
# W5A6

def calcFactorial(intCheck):
    product = 1
    if intCheck == 0 or intCheck == 1:
        return product
    else:
        for i in range(1, intCheck + 1):
            product *= i
        return product

calcFactorial(int(input("Nhập số nguyên n, trả về giai thừa của nó (n!): ")))
```

```python
# W5A7

def calc(num1, num2, operator):
    result = 0
    if operator == "+":
        result = num1 + num2
    elif operator == "-":  
        result = num1 - num2
    elif operator == "*":
        result = num1 * num2
    elif operator == "/":
        if num2 != 0:
            result = num1 / num2
        else:
            return "Lỗi không chia được cho 0"
    return round(result, 2)

calcFunction = input("Nhập lệnh gồm hai số và toán tử (+, -, *, /), giữa chúng có dấu cách. Ví dụ: 23 + 25: ")
lst = calcFunction.split()
calc(float(lst[0]), float(lst[2]), lst[1])
```

```python
# W5A8

def HammingDiff(a, b):
    diff = 0
    x = bin(a ^ b)
    for char in x:
        if char == "1":
            diff += 1
    return diff
    

x, y = map(int, input("Nhập hai số nguyên để tính khoảng cách Hamming của chúng").split())
HammingDiff(x, y)
```

```python
# W4A9

def sumDigit(intCheck):
    sum = 0
    while intCheck > 0:
        sum += intCheck % 10
        intCheck = intCheck // 10
    return sum

sumDigit(int(input("Nhập số nguyên, hàm trả về tổng các chữ số của nó:")))
```
