```python
# W6A4

mappingPair = list(input().split())
dictOut = {}

for item in mappingPair:
    tempList = item.split(":")
    if tempList[0] not in dictOut.keys():
        dictOut.update({tempList[0]: [(tempList[1])]})
    else:
        dictOut[tempList[0]].append(tempList[1])
print(dictOut)
```

```python
# W6A5

lstIn = list(map(int, input().split()))
dictOut = {"positives" : 0, "negatives": 0, "zeros": 0}

for num in lstIn:
    if num > 0:
        dictOut["positive"] += 1
    elif num < 0:
        dictOut["negative"] += 1
    else:
        dictOut["zero"] += 1

print(dictOut)
```

```python
# W6A9

dictIn1 = eval(input())
dictIn2 = eval(input())
dictOut = {}
for item in dictIn2.keys():
    if item in dictIn1.keys():
        dictIn1[item] = dictIn1[item] + dictIn2[item]
    else:
        dictIn1[item] = dictIn2[item]

for item in sorted(dictIn1.keys()):
    dictOut[item] = dictIn1[item]
print(dictOut)
```

```python
# W6A10

lstIn = list(map(int, input().split()))
k = int(input())
lstTemp = []
lstOut = []
for item in lstIn:
    if (k - item) not in lstTemp:
        lstTemp.append(item)
    else:
        if (k - item, item) not in lstOut:
            lstOut.append((k - item, item))
            lstTemp.remove(k - item)
            lstTemp.append(item)
print(sorted(lstOut))
```

```python
# W6A12

lstIn = list(map(int, input().split()))
dictTemp = {}
lstKeyTemp = []
maxTemp = 0
keyMaxTemp = 0
for item in lstIn:
    if item not in dictTemp.keys():
        dictTemp[item] = 1
        lstKeyTemp.append(item)
    else:
        dictTemp[item] += 1

lstKeyTemp = sorted(lstKeyTemp)
for item in lstKeyTemp:
    if dictTemp[item] > maxTemp:
        maxTemp = dictTemp[item]
        keyMaxTemp = item

print(keyMaxTemp)
```

```python
# W6A14
strIn1 = input()
strIn2 = input()
lstTemp = []
try:
    if strIn1 != "" and strIn2 != "":
        lstIn1 = map(int, strIn1.split())
        lstIn2 = map(int, strIn2.split())
        for item in lstIn1:
            if item in lstIn2 and item not in lstTemp:
                lstTemp.append(item)
        print(lstTemp)
    else:
        print(lstTemp)
except EOFError:
    print(lstTemp)
```

```python
# W6A17

RowCol = int(input()) # mat vuông
mat = []
for i in range(RowCol):
    rowTemp = []
    for j in range(RowCol):
        rowTemp.append(int(input()))
    mat.append(rowTemp)
mainDiag = []
subDiag = []
for i in range(RowCol):
    mainDiag.append(mat[i][i])
    subDiag.append(mat[i][RowCol - 1 - i])
print(mainDiag, subDiag)
```

```python
# W6A18
row, col, k = map(int, input().split())
mat = []
for i in range(row):
    rowTemp = []
    for j in range(col):
        rowTemp.append(int(input()))
    mat.append(rowTemp)

if k <= row:
    k = k - 1
    sumRowK = 0
    for i in range(col):
        sumRowK += mat[k][i]
    print(sumRowK)
    k = k + 1
else:
    print("Không tính được tổng hàng k lớn hơn số hàng")

if k <= col:
    k = k - 1
    sumColK = 0
    for i in range(row):
        sumColK += mat[i][k]
    print(sumColK)
    k = k + 1
else:
    print("Không tính được tổng cột k lớn hơn số cột")
```

```python
# W6A20

lstIn = list(map(int, input().split()))
k = int(input())
tempBool = True
i = 0
while tempBool:
    if i == len(lstIn):
        print(-1)
        tempBool = False
    elif lstIn[i] == k:
        print(i + 1)
        tempBool = False
    i += 1
```

```python
# W6A21
lstIn = list(input().split())
dictOut = {}
for i in range(len(lstIn)):
    if lstIn[i] not in dictOut.keys():
        dictOut.update({lstIn[i]:[i]})
    else:
        dictOut[lstIn[i]].append(i)
for item in dictOut.keys():
    dictOut[item] = tuple(dictOut[item])
print(dictOut)
```
