```python
# W8A1
pi = 3.14
class Cylinder:
    def __init__(self, r, h):
        self.r = r
        self.h = h
    def totalSurfaceArea(r, h):
        return 2*(r**2*pi) + (2*pi*r)*h
    def volume(r, h):
        return r**2*h*pi

x, y = map(int, input().split())
print(Cylinder.totalSurfaceArea(x, y))
print(Cylinder.volume(x, y))
```

```python
# W8A2
class Date:
    def __init__(self, string):
        self.day = int(string[0:2])
        self.month = int(string[3:5])
        self.year = int(string[6:])
    def Tomorrow(self):
        if self.month in {1, 3, 5, 7, 8, 10, 12}:
            if self.day > 31 or self.day < 1:
                return "INVALID"
            else:
                if self.day <= 30:
                    self.day += 1
                else:
                    self.day = 1
                    if self.month == 12:
                        self.month = 1
                        self.year += 1
                    else:
                        self.month += 1
        elif self.month in {4, 6, 9, 11}:
            if self.day > 30 or self.day < 1:
                return "INVALID"
            else:
                if self.day == 30:
                    self.day = 1
                    self.month += 1
                else:
                    self.day += 1
        elif self.month == 2:
            if self.year % 400 == 0 or (self.year % 4 == 0 and self.year % 100 != 0):
                if self.day > 29 or self.day < 1:
                    return "INVALID"
                else:
                    if self.day == 29:
                        self.day = 1
                        self.month += 1
                    else:
                        self.day += 1
            else:
                if self.day > 28 or self.day < 1:
                    return "INVALID"
                else:
                    if self.day == 28:
                        self.day = 1   
                        self.month += 1
                    else:
                        self.day += 1
        else:
            return "INVALID"    
        return f"{str(self.day).zfill(2)}/{str(self.month).zfill(2)}/{str(self.year).zfill(4)}"

d = Date(input())
print(d.Tomorrow())
                    
```

```python
# W8A3

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def eval(self):
        if self.x == self.y == 0:
            return "Gốc tọa độ"
        elif self.x == 0:
            return "Nằm trên trục tung"
        elif self.y == 0:
            return "Nằm trên trục hoành"
        else:
            return "Không nằm trên trục tung hoặc trục hoành"
    def norm(self):
        return round(((self.x)**2 + (self.y)**2)**0.5, 2)
x, y = map(int, input().split())
p = Point(x, y)
print(p.eval())
print(p.norm())
```

```python
# W8A4

class Calculator:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def add(self):
        return self.x + self.y
    def subtract(self):
        return self.x - self.y
    def multiply(self):
        return self.x * self.y
    def divide(self):
        if self.y == 0:
            return "ERROR"
        else:
            return self.x / self.y
    def power(self):
        return self.x**self.y
    def modulo(self):
        if self.y == 0:
            return "ERROR"
        else:
            return self.x % self.y
    def set_numbers(self, a, b):
        self.x = a
        self.y = b

Continue = True
x, y = map(float, input("Nhập hai số vào để tính theo thứ tự:").split())
numIn = Calculator(x, y)
while Continue:
    operator = input("Bạn muốn thực hiện tính toán như thế nào? (+, -, *, /, ^, %)?")
    if operator == "+":
        print(numIn.add())
    elif operator == "-":
        print(numIn.subtract())
    elif operator == "*":
        print(numIn.multiply())
    elif operator == "/":
        print(numIn.divide())
    elif operator == "^":
        print(numIn.power())
    elif operator == "%":
        print(numIn.modulo())
    else:
        print("Bạn đã nhập sai toán tử!")
    choice = input("Bạn muốn dừng chưa? (Y/N)")
    if choice == "Y":
        Continue = False
        print("Hẹn gặp lại lần sau!")
    else:
        x, y = map(int, input("Hãy nhập lại số nhé").split())
        numIn.set_numbers(x, y)
```

```python
# W8A5
class CartItem:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
    def subTotal(self):
        return int(self.price)*self.quantity
    def increment_quantity(self):
        self.quantity += 1
    def __repr__(self):
        return f"{self.name}, giá: {self.price}, lượng: {self.quantity}"  
class ShoppingCart:
    def __init__(self, name):
        self.name = name
        self.list = []
    def add_product(self, product_name, price):
        for item in self.list:
            if item.name == product_name:
                item.increment_quantity()
                return
        item = CartItem(product_name, price, 1)
        self.list.append(item)
    def remove_product(self, product_name):
        for item in self.list:
            if item.name == product_name:
                self.list.remove(item)
                return
    def check_is_empty(self):
        if len(self.list) == 0:
            return True
        else:
            return False
    def total_cost(self):
        sumTemp = 0
        for item in self.list:
            sumTemp += item.subTotal()
        return sumTemp
    def remove_cart(self):
        self.list = []
    def display_cart(self):
        print("Tên khách hàng: ", self.name)
        if len(self.list) == 0:
            print("Chưa có sản phẩm!")
        else:
            print(self.list)
        
cart = ShoppingCart(input("Nhập tên khách hàng: "))
is_continue = True
print('2 - Thêm một món hàng vào giỏ hàng \n'
      '3 - Loại bỏ một món hàng khỏi giỏ hàng \n'
      '4 - Kiểm tra xem giỏ hàng có trống hay không \n'
      '5 - Tính tổng giá tiền sản phẩm trong giỏ hàng \n'
      '6 - Xóa toàn bộ giỏ hàng \n'
      '7 - Biểu diễn toàn bộ giỏ hàng hiện tại \n')
while is_continue:
    choice = int(input("Nhập lựa chọn từ 2 đến 7: "))
    if choice == 2:
        product, price = input("Nhập sản phẩm và giá:").split()
        cart.add_product(product, price)
    elif choice == 3:
        cart.remove_product(input("Nhập tên món hàng bạn muốn loại bỏ"))
    elif choice == 4:
        if cart.check_is_empty():
            print("Giỏ hàng trống")
        else:
            print("Giỏ hàng có hàng")
    elif choice == 5:
        print(cart.total_cost())
    elif choice == 6:
        cart.remove_cart()
    elif choice == 7:
        cart.display_cart()
    tempBool = input("Bạn có muốn tiếp tục? (Y/N)")
    if tempBool in {"N", "n"}:
        print("Bye bye!")
        is_continue = False
```

```python
# W8A6

class Rectangle:
    def __init__(self, w, h):
        self.width = w
        self.height = h
    def area(self):
        return self.width * self.height
    def perimeter(self):
        return 2*(self.width + self.height)
    def scale(self, k):
        self.width = self.width * k
        self.height = self.height * k

w, h, k = map(float, input().split())
rec = Rectangle(w, h)
rec.scale(k)
print(rec.area(), rec.perimeter())
```

```python
# W8A7
def gcd(a, b):
    if a < 0:
        a = abs(a)
    if b < 0:
        b = abs(b)
    while a != b:
        if a > b:
            a = a - b
        else:
            b = b - a
    return a
class Fraction:
    def __init__(self, num, den):
        self.numerator = int(num)
        while den == "0":
            den = input("Nhập mẫu số khác 0: ")
        self.denominator = int(den)
        self.simplify()
    def simplify(self):
        if gcd(self.numerator, self.denominator) != 1:
            great_common_divisors = gcd(self.numerator, self.denominator)
            self.numerator = self.numerator // great_common_divisors
            self.denominator = self.denominator // great_common_divisors
    def __str__(self):
        return f"{self.numerator}/{self.denominator}"
    def __add__(self, frac2):
        sumNum = (self.numerator*frac2.denominator + self.denominator*frac2.numerator)
        sumDen = self.denominator*frac2.denominator
        return Fraction(sumNum, sumDen)
    def __sub__(self, frac2):
        subNum = (self.numerator*frac2.denominator - self.denominator*frac2.numerator)
        subDen = self.denominator*frac2.denominator
        return Fraction(subNum, subDen)
    def __mul__(self, frac2):
        mulNum = self.numerator*frac2.numerator
        mulDen = self.denominator*frac2.denominator
        return Fraction(mulNum, mulDen)
    def __truediv__(self, frac2):
        divNum = self.numerator*frac2.denominator
        divDen = self.denominator*frac2.numerator
        return Fraction(divNum, divDen)

lstIn = input().split()
frac1 = Fraction(lstIn[0], lstIn[1])
frac2 = Fraction(lstIn[3], lstIn[4])
operator = lstIn[2]
if operator == "+":
    print(frac1 + frac2)
elif operator == "-":
    print(frac1 - frac2)
elif operator == "*":
    print(frac1 * frac2)
elif operator == "/":
    print(frac1 / frac2)
```

```python
# W8A8

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)
    def __str__(self):
        return f"{self.owner}: {self.balance}"
    def deposit(self, moneyIn):
        self.balance += float(moneyIn)
    def withdraw(self, moneyOut):
        if self.balance >= float(moneyOut):
            self.balance -= float(moneyOut)

owner, initial_bal = input().split()
account = BankAccount(owner, initial_bal)
n = int(input())
for i in range(n):
    command, value = input().split()
    if command == "DEPOSIT":
        account.deposit(value)
    elif command == "WITHDRAW":
        account.withdraw(value)
print("{:.2f}".format(account.balance))
```

```python
# W8A9

class Student:
    def __init__(self, name):
        self.name = name
        self.student_data = {}
    def add_score(self, subject, score):
        self.student_data[subject] = float(score)
    def avg(self):
        count = 0
        tempSum = 0
        for item in self.student_data.keys():
            tempSum += self.student_data[item]
            count += 1
        return tempSum / count
    def rank(self):
        average = self.avg()
        if average >= 8:
            return "Excellent"
        elif average >= 6.5:
            return "Good"
        elif average >= 5:
            return "Average"
        else:
            return "Poor"
    def __str__(self):
        return f"{self.name} {"{:.2f}".format(round(self.avg(), 2))} {self.rank()}"

name = input()
student = Student(name)
n = int(input())
for i in range(n):
    subject, grade = input().split()
    student.add_score(subject, grade)
print(student)
```
